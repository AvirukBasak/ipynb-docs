### AM signal plot

````python
from math import sin
from math import cos
from math import pi as PI

import numpy as np
from matplotlib import pyplot as plt

fc = 20    # carrier freq
fm = 1     # msg freq
Ac = 5     # carrier amp
Am = 5     # msg amp

c = lambda t: Ac * sin( (2*PI * fc) * t )    # c(t) = Ac sin (2 PI fc t)
x = lambda t: Am * sin( (2*PI * fm) * t )    # x(t) = Am sin (2 PI fm t)
E = lambda t: Ac + x(t)                      # E(t) = Ac + x(t)
s = lambda t: E(t) * sin( (2*PI * fc) * t )  # s(t) = (Ac + x(t)) sin (2 PI fc t)

x_vfunc = np.vectorize(x)
c_vfunc = np.vectorize(c)
E_vfunc = np.vectorize(E)
s_vfunc = np.vectorize(s)

plt_upprbnd = 1 * PI

t_data = np.linspace(0, plt_upprbnd, 1000)

plt.plot([0, plt_upprbnd], [0, 0], color='gray')

plt.plot(t_data, c_vfunc(t_data), color='lightblue')
plt.plot(t_data, s_vfunc(t_data), color='lightgreen')
plt.plot(t_data, x_vfunc(t_data), color='orange')

# the envelope
plt.plot(t_data, E_vfunc(t_data), color='#aaaaaa')
plt.plot(t_data, -E_vfunc(t_data), color='#aaaaaa')

plt.title('DSB-FC AM signal plot')
plt.show()
````

### DSB-SC AM Signal

```
s(t) = E(t) sin (2 PI fc t)

=> s(t) = (Ac + x(t)) sin (2 PI fc t)
=> s(t) = Ac sin (2 PI fc t) + x(t) sin (2 PI fc t)

So, DSB-SC s(t) = x(t) sin (2 PI fc t)
```

````python
s = lambda t: x(t) * sin( (2*PI * fc) * t )  # s(t) = x(t) sin (2 PI fc t)
s_vfunc = np.vectorize(s)

plt.plot([0, plt_upprbnd], [0, 0], color='gray')

plt.plot(t_data, c_vfunc(t_data), color='lightblue')
plt.plot(t_data, s_vfunc(t_data), color='#33ff33')
plt.plot(t_data, x_vfunc(t_data), color='orange')

# the envelope
plt.plot(t_data, E_vfunc(t_data), color='#bbb')
plt.plot(t_data, -E_vfunc(t_data), color='#bbb')

plt.title('DSB-SC AM signal plot')
plt.show()
````

It's obvious how DSB-SC signal has a lower aplitude than DSB-FC. Hence it requires less power.

However, observe how both min and max pf `x(t)` results in same value of `s(t)`.
