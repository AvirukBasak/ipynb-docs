#### Using sigmoid only

```python
def sigmoid(x):
  return 1 / (1 + math.exp(-x))

gsimd_sg = np.zeros(shape=(512, 512, 3))
for i in range(0, 512):
    for j in range(0, 512):
        gsimd_sg[i,j] = sigmoid(imd[i,j].sum())
```

#### Using squishify
Squishify is a name I gave to a function that shifts the sigmoid curve slightly along the x-axis to get values lower than 0.5 when x >= 0

```python
def squishify(x):
    return sigmoid(x -1)

gsimd_sq = np.zeros(shape=(512, 512, 3))
for i in range(0, 512):
    for j in range(0, 512):
        gsimd_sq[i,j] = squishify(imd[i,j].sum())
```

#### Print

```python
print('squishify =\n', gsimd_sq[1:3][1:3])
print()
print('sigmoid =\n', gsimd_sg[1:3][1:3])
```

#### Plot sigmoid

```python
plt.imshow(gsimd_sg -0.25)
plt.title('sigmoid')
plt.show()
```
