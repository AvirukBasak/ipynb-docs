### 3D plot using matplotlib

```python
dda = np.array([np.random.random(100) for x in range(0, 3)]) * 100
ax = plt.axes(projection='3d')
ax.scatter3D(dda[0], dda[1], dda[2], color='red')
plt.title('3D scatter plot')
plt.show()
```

### Accessing multi-d np arrays

There are two syntaxes for accessing an np milti-dimensional array

```python
def equals():
    for i in range(3):
        for j in range(100):
            if dda[i,j] != dda[i][j]:
                return False
    return True
equals()
```

### Math functions with matplotlib

```python
import math

data_x = np.arange(0, 2 * math.pi, 0.01)

sin_vfunc = np.vectorize(lambda x: math.sin(x))
cos_vfunc = np.vectorize(lambda x: math.cos(x))

plt.plot([0, 2 * math.pi], [0, 0], color='gray')    # ref line y = 0
plt.plot(data_x, sin_vfunc(data_x), color='blue')   # y = sin(x)
plt.plot(data_x, cos_vfunc(data_x), color='red')    # y = cos(x)

plt.title('sin(x) and cos(x)')
plt.show()
```

```python
tan_vfunc = np.vectorize(lambda x: math.tan(x))

plt.plot([0, 2 * math.pi], [0, 0], color='gray')                          # ref line y = 0
plt.plot(data_x, tan_vfunc(data_x), color='red')                          # y = tan(x)
plt.ylim(-5, 5)
plt.title('tan(x)')
plt.show()


plt.plot([0, 2 * math.pi], [0, 0], color='gray')                          # ref line y = 0
plt.plot(data_x, sin_vfunc(data_x) / cos_vfunc(data_x), color='green')    # y = sin(x) / cos(x)
plt.ylim(-5, 5)
plt.title('tan(x)')
plt.show()
```

### Using numpy for plotting math functions

Note that `np.arange` infers the number of data points from the passed step size, whereas `np.linspace` infers the step size from the passed number of data points.

```python
data_x = np.linspace(0, 2 * math.pi, 100)
plt.plot([0, 2 * math.pi], [0, 0], color='gray')
plt.plot(data_x, np.tan(data_x), color='green')
plt.title('np.tan(x)')
plt.ylim(-5, 5)
plt.show()
```
