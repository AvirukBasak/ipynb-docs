### PM signal plot

```python
from math import sin
from math import cos
from math import pi as PI

import numpy as np
from matplotlib import pyplot as plt

fc = 10    # carrier freq
fm = 1     # msg freq
Ac = 5     # carrier amp
Am = 5     # msg amp

# phase sensitivity
kp = 1

c = lambda t: Ac * sin( (2*PI * fc) * t )              # c(t) = Ac sin (2 PI fc t)
x = lambda t: Am * sin( (2*PI * fm) * t )              # x(t) = Am sin (2 PI fm t)
phi_i = lambda t: kp * x(t)                            # phi_i(t) = kp x(t)
s = lambda t: Ac * sin( (2*PI * fc) * t  + phi_i(t) )  # s(t) = Ac sin (2 PI fc t + phi_i(t))

x_vfunc = np.vectorize(x)
c_vfunc = np.vectorize(c)
phi_i_vfunc = np.vectorize(phi_i)
s_vfunc = np.vectorize(s)

plt_upprbnd = 1 * PI

t_data = np.linspace(0, plt_upprbnd, 1000)

plt.plot([0, plt_upprbnd], [0, 0], color='gray')

plt.plot(t_data, c_vfunc(t_data), color='#bbbbff')
plt.plot(t_data, s_vfunc(t_data), color='#4444')
plt.plot(t_data, x_vfunc(t_data), color='#ff5555')

plt.title('PM signal plot')
plt.show()
```
